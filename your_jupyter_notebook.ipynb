```python
# --- 0. JUPYTER SETUP ---
# This "magic command" ensures that plots are displayed directly in the notebook.
%matplotlib inline

# --- 1. IMPORTS ---
import os
import pandas as pd
import numpy as np
from pathlib import Path  # Import Path for robust path handling
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import KaplanMeierFitter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.metrics import roc_auc_score
import shap
import pickle

# --- 2. CONFIGURATION & DYNAMIC PATHS ---
try:
    NOTEBOOK_DIR = Path().resolve()
except NameError:
    NOTEBOOK_DIR = Path.cwd()

DATA_DIR = NOTEBOOK_DIR / 'data'
MODELS_DIR = NOTEBOOK_DIR / 'models'
# --- Use the correct Kaggle filenames ---
KAGGLE_DATA_FILENAME = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'
CLEANED_DATA_PATH = DATA_DIR / 'cleaned_telco_customers.csv'  # <-- Correct output name
MODEL_PATH = MODELS_DIR / 'churn_model_telco_v1.pkl'

KAGGLE_DATA_PATH = DATA_DIR / KAGGLE_DATA_FILENAME

# Business assumptions
AVG_CLV_INCREASE_PER_RETENTION = 800
COST_PER_RETENTION_OFFER = 50

# --- 3. ANALYSIS & MODELING FUNCTIONS ---

def load_and_clean_kaggle_data():
    """Loads and cleans the Telco Customer Churn dataset from Kaggle."""
    print("Step 1: Loading and cleaning data from Kaggle...")
    
    if not KAGGLE_DATA_PATH.is_file():
        print(f"--- ERROR: The file was not found at {KAGGLE_DATA_PATH} ---")
        print("Please ensure 'WA_Fn-UseC_-Telco-Customer-Churn.csv' is inside a 'data' folder next to your notebook.")
        return None

    df = pd.read_csv(KAGGLE_DATA_PATH)

    # --- Real-World Data Cleaning ---
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
    df.dropna(subset=['TotalCharges'], inplace=True)
    df['churn_label'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

    # --- Feature Engineering ---
    df.rename(columns={'tenure': 'tenure_months'}, inplace=True)
    df['tenure_bucket'] = pd.cut(
        df['tenure_months'], 
        bins=[0, 12, 24, 48, 72], 
        labels=['0-1 Year', '1-2 Years', '2-4 Years', '4+ Years']
    )
    df.drop(columns=['Churn'], inplace=True)

    # âš¡ NEW: Save cleaned file so Streamlit dashboard can load it
    df.to_csv(CLEANED_DATA_PATH, index=False)
    print(f"ðŸ’¾ Cleaned data saved to {CLEANED_DATA_PATH}")

    print(f"âœ… Kaggle data loaded and cleaned successfully.")
    return df

def perform_eda_and_visuals(df):
    """Performs EDA and displays plots directly in the notebook."""
    if df is None: return
    print("\nStep 2: Performing EDA and generating visuals...")
    
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # Churn by Contract Type
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='Contract', hue='churn_label', palette='viridis')
    plt.title('Churn by Contract Type', fontsize=16)
    plt.show()

    # Churn by Tenure Bucket
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='tenure_bucket', hue='churn_label', palette='plasma')
    plt.title('Churn by Tenure', fontsize=16)
    plt.show()

    # Survival Analysis
    kmf = KaplanMeierFitter()
    kmf.fit(df['tenure_months'], event_observed=df['churn_label'])
    plt.figure(figsize=(10, 6))
    kmf.plot(label='Survival Probability')
    plt.title('Customer Survival Curve (Telco Data)', fontsize=16)
    plt.xlabel('Tenure (Months)')
    plt.ylabel('Probability of Retention')
    plt.grid(True)
    plt.show()

    print(f"âœ… EDA plots displayed above.")

def train_and_evaluate_model(df):
    """Trains, evaluates, and saves the predictive model."""
    if df is None: return
    print("\nStep 3: Training and evaluating the model...")
    
    df_model = df.copy()
    TARGET = 'churn_label'
    
    # Features from the Kaggle Dataset
    numeric_features = ['tenure_months', 'MonthlyCharges', 'TotalCharges']
    categorical_features = [
        'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 
        'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
        'Contract', 'PaperlessBilling', 'PaymentMethod'
    ]
    
    # Drop columns that are not features for the model
    df_model.drop(columns=['customerID', 'tenure_bucket'], inplace=True, errors='ignore')

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_features),
            ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features)
        ], remainder='passthrough'
    )

    X = df_model.drop(TARGET, axis=1)
    y = df_model[TARGET]
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
    
    # Handle imbalance
    scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]
    
    xgb_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', xgb.XGBClassifier(
            use_label_encoder=False, 
            eval_metric='logloss', 
            random_state=42, 
            scale_pos_weight=scale_pos_weight
        ))
    ])
    xgb_pipeline.fit(X_train, y_train)

    # Evaluation
    y_pred_proba = xgb_pipeline.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_pred_proba)
    print(f"ðŸ“Š Model AUC-ROC: {auc:.4f}")
    
    # SHAP Explainability
    print("\nGenerating SHAP feature importance plot...")
    transformed_X_test = pd.DataFrame(
        xgb_pipeline.named_steps['preprocessor'].transform(X_test),
        columns=xgb_pipeline.named_steps['preprocessor'].get_feature_names_out()
    )
    explainer = shap.TreeExplainer(xgb_pipeline.named_steps['classifier'])
    shap_values = explainer.shap_values(transformed_X_test)
    
    shap.summary_plot(shap_values, transformed_X_test, show=True)
    
    # Save the model
    MODELS_DIR.mkdir(exist_ok=True)
    with open(MODEL_PATH, 'wb') as f:
        pickle.dump(xgb_pipeline, f)
    print(f"âœ… Model trained and saved to '{MODEL_PATH}'")

# --- 4. EXECUTION BLOCK ---
print("--- Running Full Churn Analysis Pipeline with Kaggle Data ---")

# Create directories if they don't exist
DATA_DIR.mkdir(exist_ok=True)
MODELS_DIR.mkdir(exist_ok=True)

# Run the full pipeline
cleaned_df = load_and_clean_kaggle_data()
perform_eda_and_visuals(cleaned_df)
train_and_evaluate_model(cleaned_df)

print("\n--- âœ… ANALYSIS PIPELINE COMPLETE ---")
```
